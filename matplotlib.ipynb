```python
x = np.arange(100)

years = [2020 + x for x in range(6)]
income = [53, 55, 59, 54, 63, 65]
income_list = list(range(50, 72, 2))

fig, axis = plt.subplots(2, 2, figsize=(14, 12))

axis[0, 0].plot(years, income)
axis[0, 0].set_title("Arsen's Income")
axis[0, 0].set_xlabel("Years")
axis[0, 0].set_ylabel("Income")
axis[0, 0].set_yticks(income_list, [f"${x}k" for x in income_list])

axis[0, 1].plot(x, np.sin(x))
axis[0, 1].set_title("Sinus Func")
axis[0, 1].set_xlabel("Degrees")
axis[0, 1].set_ylabel("Values")

axis[1, 0].plot(x, np.cos(x))
axis[1, 0].set_title("Cosinus Func")
axis[1, 0].set_xlabel("Degrees")
axis[1, 0].set_ylabel("Values")

axis[1, 1].plot(x, np.log2(x))
axis[1, 1].set_title("Logarithm Func")
axis[1, 1].set_xlabel("Numbers")
axis[1, 1].set_ylabel("Values")

fig.suptitle("Four Plots")
```
